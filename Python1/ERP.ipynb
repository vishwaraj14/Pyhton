#Merging
# For the intersection of two dataframe values under all the common  columns should be same 
#


```python
pune=pd.read_excel('CDAC_DataBook.xlsx',sheet_name='Pune')
```

```python
mum=pd.read_excel('CDAC_DataBook.xlsx',sheet_name='Mumbai')
```

```python
pune
```

```python
mum
```

```python
pd.merge(pune,mum) #For the intersection of two dataframe values under all the common  columns should be same 
```

```python
pd.merge(pune,mum, on='Subject') ## here  how is  default as  inner
```

```python
# for pairing it will be repeat  # 1 cousre 2 time eg pyhton in pune 2 times , but in mumbai it is 1 , so mumbai will print 2 times   - for pairing 
```

```python
# FOR UINION  
pd.merge(pune,mum, on='Subject',how ='outer')
```

```python
pd.merge(pune,mum,on=['Subject','Grade']) # 
```

```python
 staked=pd.melt(mum,id_vars="Name",var_name="Subject")
 print(staked)
```

```python
 staked1=pd.melt(pune,id_vars="Name",var_name="Subject")
 print(staked)
```

```python
pd.merge(staked,staked1,on=['Subject'],how='inner')
```
